```python
player = pd.DataFrame([])
```

```python
player['AB'] = bats.groupby(['batter.id']).count()['batter.name'].sort_values(ascending=False)
```

```python
player['RE'] = bats.groupby(['batter.id']).sum(numeric_only=True)['RE'].sort_values(ascending=False)
```

```python
player = pd.merge(player, profile, left_on=['batter.id'], right_on=['id'], how='left')
```

```python
player[['id', 'name', 'team_code', 'AB', 'RE']].sort_values('RE', ascending=False)
```

```python
player.sort_values('RE', ascending=False).groupby(['team_code'], as_index=False).sum(numeric_only=True)[['team_code', 'AB', 'RE']].sort_values('RE', ascending=False)
```
